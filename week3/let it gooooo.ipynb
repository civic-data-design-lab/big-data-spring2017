```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from itertools import cycle, islice

# This line lets us plot on our ipython notebook
%matplotlib inline
```

```python
#Import df_voters
df_voters = pd.read_table('data/CENTRE_FVE_20170123.csv', sep=',', low_memory=False)
df_voters
```

```python
# View the unique values in column 12
df_voters['12'].unique()
```

```python
# Your code here
grouped = df_voters.groupby('12') # Group by party registrations
summed = grouped['12'].count() # Get counts of each group
summed.sort_values(inplace=True, ascending=False)
df_summed = summed.to_frame()
df_summed.columns = ['NUMBER']
df_summed.index.names = ['PARTY']
df_summed
```

```python
df_summed[(df_summed.index != 'D') & (df_summed.index != 'R')]
```

```python
parties = ['Democrat', 'Republican', 'Other']
values = np.array([df_summed[(df_summed.index == 'D')].NUMBER.item(),df_summed[(df_summed.index == 'R')].NUMBER.item(),df_summed[(df_summed.index != 'D') & (df_summed.index != 'R')].NUMBER.sum()])
df_parties = pd.DataFrame(values,index=parties,columns=['voters'])
df_parties
```

```python
df_parties.plot.bar(color='BRY')
plt.ylabel('Registered Voters')
plt.xlabel('Party')
plt.title('Party affiliation')
plt.show()

```

```python
df_prec42 = df_voters[df_voters['27'] == 42]
df_prec42.head()
```

```python
grouped42 = df_prec42.groupby('12') # Group by party registrations
summed42 = grouped42['12'].count() # Get counts of each group
summed42.sort_values(inplace=True, ascending=False)
df_summed42 = summed42.to_frame()
df_summed42.columns = ['NUMBER']
df_summed42.index.names = ['PARTY']
df_summed42
```

```python
parties = ['Democrat', 'Republican', 'Other']
values = np.array([df_summed42[(df_summed42.index == 'D')].NUMBER.item(),df_summed42[(df_summed42.index == 'R')].NUMBER.item(),df_summed42[(df_summed42.index != 'D') & (df_summed42.index != 'R')].NUMBER.sum()])
df_parties42 = pd.DataFrame(values,index=parties,columns=list('N'))
df_parties42   
```

```python
df_results = pd.read_table('data/centrecounty_precinct_results_nov2016.txt', sep=',')
df_results.head()
```

```python
df_res_prec42 = df_results[df_results['PrecNo'] == 42]
```

```python
df_respres_p42 = df_res_prec42.loc [df_res_prec42['Contest'] == 'PRESIDENTIAL ELECTORS']
df_respres_p42.head()
```

```python
df_respres_p42.groupby('Candidate').Count.sum()
```

```python
grouped_res42 = df_respres_p42.groupby('Candidate')
summed_res42 = grouped_res42['Count'].sum()
summed_res42
```

```python
grouped_res42 = df_respres_p42.groupby('Candidate') # Group by party registrations
summed_res42 = grouped_res42['Count'].sum() # Get counts of each group
summed_res42.sort_values(inplace=True, ascending=False)
df_summed_res42 = summed_res42.to_frame()
df_summed_res42.columns = ['VOTES']
df_summed_res42.index.names = ['CANDIDATE']
df_summed_res42
```

```python
parties = ['Democrat','Republican','Other']
values_res = np.array([df_summed_res42[(df_summed_res42.index == 'HILLARY CLINTON,  PRESIDENT')].VOTES.item(),df_summed_res42[(df_summed_res42.index == 'DONALD J TRUMP,  PRESIDENT')].VOTES.item(),df_summed_res42[(df_summed_res42.index != 'HILLARY CLINTON,  PRESIDENT') & (df_summed_res42.index != 'DONALD J TRUMP,  PRESIDENT')].VOTES.sum()])
df_results42 = pd.DataFrame(values_res,index=parties,columns=list('V'))
df_results42   
```

```python
df_prec42_diff = pd.concat([df_results42, df_parties42], axis=1)
df_prec42_diff
```

```python
df_prec42_diff.plot.bar(color = 'BR')
plt.title('Precinct 42 Registered voters and results')
plt
```

```python
df_rhillary = df_results[(df_results['Contest'] == 'PRESIDENTIAL ELECTORS') & (df_results['Candidate'] == 'HILLARY CLINTON,  PRESIDENT')]
df_rhillary
```

```python
df_plothillary = df_rhillary[['PrecNo','PctCnt']]
df_plothillary
```
